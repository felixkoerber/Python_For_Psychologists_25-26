### Group By

If we have different groups within our data - in our case for example we have some congruent and some incongruent trials - we can use `groupby` to get some grouped summary statistics for example, mean accuracy per group.

```python
df.groupby("congruency")["correct"].mean()
```

```python
df.groupby("congruency")["rt"].mean()
```

instead of computing the mean we can also use groupby to plot grouped histograms:

```python
df.groupby("congruency")["rt"].hist()
```

# Visualisation with seaborn

Let's also go over some data visualisation using the widely used package [seaborn](https://seaborn.pydata.org)

```python
import seaborn as sns
```

Load an example dataset from seaborn

```python
# Load an example dataset
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

The default for `relplot` are scatterplots as you can see. But you can easily change it to lineplots using the `kind` argument. Lets try it on a new dataset that is better fitted for a lineplot: timeseries data

```python
dots = sns.load_dataset("dots")
```

```python
dots
```

```python
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

Including statistical estimates: means and confidence intervals

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

You can also easily include a linear regression model and its uncertainty using `lmplot`:

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

<div class="alert alert-block alert-info">
<b>Seaborn plots:</b> As you can see, these plots always follow a very similar structure: define what kind of plot you want (check documentation for available types), define what should be on the x-axis and what should be on the y-axis, provide name of the dataframe.
</div>
