Today we will have a look at data visualisation with matplotlib - the backbone of plotting in python. Check out [seaborn](https://seaborn.pydata.org) a "wrapper" over matplotlib with a few "quality of life" convenience functions and a ton of examples.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt  # pyplot is the base plotting module in python
```

Lets generate some data

```python
a = np.random.random(100)  #
b = np.random.randn(100)  # 
```

Matplotlib has a lot of low-level straight forward plotting functions, e.g., histograms

```python
plt.hist(b)
plt.hist(a)
```

As you can see, `plt` will plot both arrays on the same axis automatically assigning different colors.

We can also plot lines very easily:

```python
plt.plot(b)
plt.plot(a)
```

```python
plt.plot(b,a)
```

this doesn't really look like it makes much sense... So how about we first sort the values?

```python
plt.plot(sorted(a), sorted(b))  # a gegen b in sortierter Reihenfolge
```

You can also draw lines by providing specific coordinates:

```python
plt.plot((0, 1), (2, 3)) #(x1,x2),(y1,y2)
```

You can add different styles to `plt` functions like `hist`, `plt` and `scatter` using arguments. For example we can add styles to the axes: 

```python
plt.plot(b, 'r+')  # red cross
plt.plot(a, color="blue", marker="o", linewidth=0)  # blue circle

plt.ylabel("nothing")
plt.xlabel("hello")
```

Usually, when we want to create more elaborate plots we first create the empty axes for our plot and subsequently fill it with the data of interest. In that case you first create the axes like so:

```python
ax = plt.axes()
ax
```

Which returns an empty set of axes.

then you can call the different plotting functions and styles directly to the axes.

```python
ax = plt.axes()  # leere Axe

ax.plot(a - 5)  # line plot
ax.hist((b * 20) + 40)  # histogram (skaliert)
ax.scatter(20 + a * 20, b - 10, c="r")  # scatterplot

ax.set_ylabel("nothing")
ax.set_xlabel("all")
```

You can add a legend by defining `labels`. In the end you can call `plt.legend()` to draw the legend.

```python
ax = plt.axes()
ax.scatter(a, b, color='r', label="red dots")
ax.scatter(b, a, color='b', label="blue dots")
plt.legend()
```
